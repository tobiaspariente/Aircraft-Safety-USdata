```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = df_clean
```

## 2- Exploration of simple relationships

### 2.1- Investigation.Type

Objective: Identify the most common types of investigations to understand if certain investigation types are more frequently associated with severe accidents.

```python
# Count occurrences of each investigation type
investigation_counts = df['Investigation.Type'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
investigation_counts.plot(kind='bar')
plt.title('Number of Accidents by Investigation Type')
plt.xlabel('Investigation Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### 2.2- Event.Date

Objective: Examine trends in accident frequency over time to identify any patterns or trends associated with specific periods.

```python
# Convert Event.Date to datetime
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')

# Extract year from Event.Date
df['Year'] = df['Event.Date'].dt.year

# Count the number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()

# Plotting the results
plt.figure(figsize=(12, 6))
accidents_per_year.plot(kind='line', marker='o')
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

### 2.3- Injury.Severity

Objective: Analyze the distribution of injury severities to understand which types of injuries are most commonly associated with accidents.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Injury.Severity', order=df['Injury.Severity'].value_counts().index)
plt.title('Injury Severity in Accidents')
plt.xlabel('Injury Severity')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right', fontsize=10)  # Rotate the x-axis labels
plt.show()
```

### 2.4- Aircraft.damage

## 3- New columns

```python
df['Total.Occupants'] = (df['Total.Fatal.Injuries'] + 
                         df['Total.Serious.Injuries'] + 
                         df['Total.Minor.Injuries'] + 
                         df['Total.Uninjured'])

df['Proportion.Affected'] = (df['Total.Fatal.Injuries'] + 
                             df['Total.Serious.Injuries'] + 
                             df['Total.Minor.Injuries']) / df['Total.Occupants']

df[['Make', 'Model', 'Total.Occupants', 'Proportion.Affected']].head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Proportion.Affected'], bins=30, kde=True)
plt.title('Distribution of Affected Occupants Proportion')
plt.xlabel('Proportion of Occupants Affected')
plt.ylabel('Frequency')
plt.show()
```

Explanation: The histogram shows the distribution of the proportion of affected occupants in accidents. Most values are near 0, meaning few occupants were affected in many cases. The KDE line highlights the overall trend, showing how these proportions are distributed across all incidents.

```python
df_model_avg = df.groupby(['Make', 'Model'])['Proportion.Affected'].mean().reset_index()

df_model_avg_sorted = df_model_avg.sort_values(by='Proportion.Affected')

plt.figure(figsize=(10, 8))
sns.barplot(x='Proportion.Affected', y='Model', data=df_model_avg_sorted.head(5), palette='Blues_r')
plt.title('Top 5 Safest Aircraft Models (Lowest Proportion of Affected Occupants)')
plt.xlabel('Average Proportion of Affected Occupants')
plt.ylabel('Aircraft Model')
plt.show()
```

```python
df_model_avg_sorted = df_model_avg.sort_values(by='Proportion.Affected', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x='Proportion.Affected', y='Model', data=df_model_avg_sorted.head(5), palette='Reds_r')
plt.title('Top 5 Least Safe Aircraft Models (Highest Proportion of Affected Occupants)')
plt.xlabel('Average Proportion of Affected Occupants')
plt.ylabel('Aircraft Model')
plt.show()
```
